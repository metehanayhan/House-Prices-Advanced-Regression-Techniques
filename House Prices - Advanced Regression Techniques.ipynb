### Data Dictionary
- MSSubClass: Satışa konu olan konut türünü tanımlar. Örneğin, "20" değeri, 1946'dan yeni tek katlı evleri ifade eder.
- MSZoning: Satışın genel imar sınıflandırmasını belirtir. Örneğin, "A" tarımsal bölgeyi ifade eder.
- LotFrontage: Mülke bağlı olan caddeye bitişik cephe genişliği (lineer feet olarak).
- LotArea: Mülkün yüzölçümü (square feet olarak).
- Street: Mülke erişim sağlayan yol türü. Örneğin, "Grvl" çakıl, "Pave" asfalt demektir.
- Alley: Mülke erişim sağlayan arka sokak türü. "NA" değeri, arka sokak erişimi olmadığını belirtir.
- LotShape: Mülkün genel şekli. Örneğin, "Reg" düzenli, "IR3" düzensiz demektir.
- LandContour: Mülkün düzlüğü veya eğimi. Örneğin, "Lvl" neredeyse düz, "HLS" tepe üstü eğimli demektir.
- Utilities: Mülkte mevcut olan altyapı türleri. Örneğin, "AllPub" tüm kamu hizmetlerinin mevcut olduğunu ifade eder.
- LotConfig: Mülkün yerleşim şekli. Örneğin, "Inside" içte kalan, "Corner" köşe arsa demektir.
- LandSlope: Mülkün eğimi. "Gtl" hafif eğim, "Sev" şiddetli eğim anlamına gelir.
- Neighborhood: Mülkün Ames şehir sınırları içindeki fiziksel konumu. Örneğin, "OldTown" eski şehir merkezi demektir.
- Condition1: Mülkün çeşitli koşullara olan yakınlığı. "Artery" ana caddeye bitişik, "PosN" parka veya yeşil alan gibi olumlu bir dış faktöre yakın anlamına gelir.
- BldgType: Konut tipi. "1Fam" tek ailelik müstakil ev, "Duplx" ikiz ev demektir.
- HouseStyle: Konut stili. "1Story" tek katlı, "2Story" iki katlı ev anlamına gelir.
- OverallQual: Evin genel malzeme ve bitiş kalitesi, 1 (çok kötü) ile 10 (çok mükemmel) arasında derecelendirilir.
- OverallCond: Evin genel durumu, 1 (çok kötü) ile 10 (çok mükemmel) arasında derecelendirilir.
- YearBuilt: Evin inşa edildiği yıl.
- YearRemodAdd: Evin tadilat veya ekleme yapılan yılı (eğer tadilat yapılmamışsa, inşa yılı ile aynı olabilir).
- RoofStyle: Çatı tipi. Örneğin, "Gable" kırma çatı, "Flat" düz çatı demektir.
- RoofMatl: Çatı malzemesi. "CompShg" kompozit şingıl, "Metal" metal demektir.
- Exterior1st ve Exterior2nd: Evin dış kaplama malzemesi. Örneğin, "VinylSd" vinil kaplama, "Stone" taş kaplama anlamına gelir.
- MasVnrType: Duvar kaplama türü. Örneğin, "BrkFace" tuğla kaplama, "Stone" taş kaplama anlamına gelir.
- MasVnrArea: Duvar kaplamasının alanı (square feet olarak).
- ExterQual: Dış malzemenin kalitesi, "Ex" (mükemmel) ile "Po" (kötü) arasında derecelendirilir.
- ExterCond: Dış malzemenin mevcut durumu, "Ex" (mükemmel) ile "Po" (kötü) arasında derecelendirilir.
- Foundation: Temel tipi. Örneğin, "PConc" dökme beton temel demektir.
- BsmtQual: Bodrum yüksekliği, "Ex" (mükemmel, 100 inçten büyük) ile "Po" (kötü, 70 inçten küçük) arasında derecelendirilir.
- BsmtCond: Bodrumun genel durumu, "Ex" (mükemmel) ile "Po" (kötü) arasında derecelendirilir.
- BsmtExposure: Bodrumda yürüyüş veya bahçe seviyesi duvarlar. "Gd" iyi, "No" ise hiçbir maruz kalma yok anlamına gelir.
- BsmtFinType1 ve BsmtFinType2: Bodrumda bitmiş alanların derecesi, "GLQ" (iyi yaşam alanı) ile "Unf" (bitmemiş) arasında derecelendirilir.
- BsmtFinSF1 ve BsmtFinSF2: Birinci ve ikinci tür bitmiş bodrum alanının alanı (square feet olarak).
- BsmtUnfSF: Bodrumdaki bitmemiş alanın büyüklüğü.
- TotalBsmtSF: Bodrumun toplam alanı (square feet olarak).
- Heating: Isıtma türü. Örneğin, "GasA" gazlı zorlamalı sıcak hava fırını anlamına gelir.
- HeatingQC: Isıtma kalitesi ve durumu, "Ex" (mükemmel) ile "Po" (kötü) arasında derecelendirilir.
- CentralAir: Merkezi klima olup olmadığını belirtir. "Y" evet, "N" hayır anlamına gelir.
- Electrical: Elektrik sistemi. "SBrkr" standart devre kesiciler ve Romex kablolama anlamına gelir.
- 1stFlrSF: Birinci katın alanı (square feet olarak).
- 2ndFlrSF: İkinci katın alanı (square feet olarak).
- LowQualFinSF: Düşük kaliteli bitmiş alan (tüm katlar dahil).
- GrLivArea: Üst zemin (yer üstü) yaşam alanı (square feet olarak).
- BsmtFullBath: Bodrumda tam banyo sayısı.
- BsmtHalfBath: Bodrumda yarım banyo sayısı.
- FullBath: Yer üstünde tam banyo sayısı.
- HalfBath: Yer üstünde yarım banyo sayısı.
- Bedroom: Yer üstündeki yatak odası sayısı (bodrum yatak odaları dahil değildir).
- Kitchen: Yer üstündeki mutfak sayısı.
- KitchenQual: Mutfak kalitesi, "Ex" (mükemmel) ile "Po" (kötü) arasında derecelendirilir.
- TotRmsAbvGrd: Yer üstünde bulunan toplam oda sayısı (banyolar dahil değildir).
- Functional: Evin işlevselliği, "Typ" (tipik) ile "Sal" (sadece kurtarma) arasında derecelendirilir.
- Fireplaces: Şömine sayısı.
- FireplaceQu: Şömine kalitesi, "Ex" (mükemmel) ile "Po" (kötü) arasında derecelendirilir.
- GarageType: Garaj konumu. "Attchd" eve bağlı garaj, "Detchd" evden bağımsız garaj demektir.
- GarageYrBlt: Garajın inşa edildiği yıl.
- GarageFinish: Garajın iç bitirme durumu, "Fin" bitmiş, "Unf" bitmemiş anlamına gelir.
- GarageCars: Garajın araba kapasitesi.
- GarageArea: Garajın alanı (square feet olarak).
- GarageQual: Garajın kalitesi, "Ex" (mükemmel) ile "Po" (kötü) arasında derecelendirilir.
- GarageCond: Garajın durumu, "Ex" (mükemmel) ile "Po" (kötü) arasında derecelendirilir.
- PavedDrive: Asfalt yol olup olmadığını belirtir. "Y" evet, "N" hayır anlamına gelir.
- WoodDeckSF: Ahşap veranda alanı (square feet olarak).
- OpenPorchSF: Açık veranda alanı (square feet olarak).
- EnclosedPorch: Kapalı veranda alanı (square feet olarak).
- 3SsnPorch: 3 mevsim veranda alanı (square feet olarak).
- ScreenPorch: Ekranlı veranda alanı (square feet olarak).
- PoolArea: Havuz alanı (square feet olarak).
- PoolQC: Havuzun kalitesi, "Ex" (mükemmel) ile "Po" (kötü) arasında derecelendirilir.
- Fence: Çit kalitesi, "GdPrv" iyi gizlilik, "MnWw" minimum ahşap/çelik anlamına gelir.
- MiscFeature: Diğer özellikler, "Elev" asansör, "TenC" tenis kortu anlamına gelir.
- MiscVal: Diğer özelliklerin değeri.
- MoSold: Satışın gerçekleştiği ay.
- YrSold: Satışın gerçekleştiği yıl.
- SaleType: Satış türü, "WD" garanti belgesi anlamına gelir.
- SaleCondition: Satışın koşulları, "Normal" tipik satış anlamına gelir.

### Filling Missing Data and Data Conversion

```python
# KNN Imputer ile doldurma= benzer ev özelliklerinin verisi ile dolduruldu en yakın komşu..
knn_imputer = KNNImputer(n_neighbors=3)
df[['LotFrontage','GarageYrBlt', 'GarageCars', 'GarageArea']] = knn_imputer.fit_transform(df[['LotFrontage','GarageYrBlt', 'GarageCars', 'GarageArea']])
```

```python
df = df.drop(columns=['PoolQC', 'Fence', 'MiscFeature','Alley'])
```

```python
df[['MSZoning','Utilities', 'Exterior1st', 'Exterior2nd', 'MasVnrType', 'Electrical', 'BsmtFullBath', 'BsmtHalfBath', 'KitchenQual', 'Functional', 'SaleType']] = df[['MSZoning','Utilities', 'Exterior1st', 'Exterior2nd', 'MasVnrType', 'Electrical', 'BsmtFullBath', 'BsmtHalfBath', 'KitchenQual', 'Functional', 'SaleType']].apply(lambda x: x.fillna(x.mode()[0]))
```

```python
df['MasVnrArea'] = df.groupby('MasVnrType')['MasVnrArea'].transform(lambda x: x.fillna(x.median()))
```

```python
label_encoders = {}
categorical_cols = ['BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2']

for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))
    label_encoders[col] = le

knn_imputer = KNNImputer(n_neighbors=3)
df[['BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2','BsmtUnfSF','TotalBsmtSF','BsmtFinSF1','BsmtFinSF2']] = knn_imputer.fit_transform(df[['BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2','BsmtUnfSF','TotalBsmtSF','BsmtFinSF1','BsmtFinSF2']])
```

```python
imputer_mode = SimpleImputer(strategy='most_frequent')

df[['FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond']] = imputer_mode.fit_transform(df[['FireplaceQu', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond']])
```

```python
# Feature Engineers
df['AgeAtSale'] = df['YrSold'] - df['YearBuilt']
df['RemodelAge'] = df['YearRemodAdd'] - df['YearBuilt']
df['TotalBathrooms'] = df['FullBath'] + 0.5 * df['HalfBath'] + df['BsmtFullBath'] + 0.5 * df['BsmtHalfBath']
```

```python
categorical_cols = [
    'MSSubClass', 'MSZoning', 'Street', 'Utilities',
    'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2', 'BldgType',
    'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st', 'Exterior2nd', 'MasVnrType',
    'Foundation', 'Electrical', 'CentralAir', 'Functional', 'GarageType', 'GarageFinish',
    'PavedDrive', 'SaleType', 'Heating', 'SaleCondition'
]

df = pd.get_dummies(df, columns=categorical_cols, prefix=categorical_cols)
```

```python
df['LotShape'] = df['LotShape'].map({'Reg': 1, 'IR1': 2, 'IR2': 3, 'IR3': 4})
df['LandContour'] = df['LandContour'].map({'Lvl': 1, 'Bnk': 2, 'HLS': 3, 'Low': 4})
df['ExterQual'] = df['ExterQual'].map({'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1})
df['ExterCond'] = df['ExterCond'].map({'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1})
df['HeatingQC'] = df['HeatingQC'].map({'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1})
df['KitchenQual'] = df['KitchenQual'].map({'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1})
df['FireplaceQu'] = df['FireplaceQu'].map({'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1, 'NA': 0})
df['GarageQual'] = df['GarageQual'].map({'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1, 'NA': 0})
df['GarageCond'] = df['GarageCond'].map({'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1, 'NA': 0})
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
pd.set_option("display.max_columns",100)

from sklearn.linear_model import LinearRegression,SGDRegressor,Ridge,Lasso,ElasticNet
from sklearn.neighbors import KNeighborsRegressor, RadiusNeighborsRegressor
from sklearn.ensemble import GradientBoostingRegressor,AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor, plot_tree, ExtraTreeRegressor
from xgboost import XGBRegressor
from sklearn.svm import SVR

from sklearn.neural_network import MLPRegressor

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error

from sklearn.preprocessing import StandardScaler


def algo_test(x,y):
        #Bütün modelleri tanımlıyorum
        L=LinearRegression()
        R=Ridge()
        Lass=Lasso()
        E=ElasticNet()
        sgd=SGDRegressor()
        ETR=ExtraTreeRegressor()
        GBR=GradientBoostingRegressor()
        kn=KNeighborsRegressor()
        rkn=RadiusNeighborsRegressor(radius=1.0)
        ada=AdaBoostRegressor()
        dt=DecisionTreeRegressor()
        xgb=XGBRegressor()
        svr=SVR()
        mlp_regressor = MLPRegressor()

       
        
        algos=[L,R,Lass,E,sgd,ETR,GBR,ada,kn,dt,xgb,svr,mlp_regressor]
        algo_names=['Linear','Ridge','Lasso','ElasticNet','SGD','Extra Tree','Gradient Boosting',
                    'KNeighborsRegressor','AdaBoost','Decision Tree','XGBRegressor','SVR','mlp_regressor']
        
        x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=.20,random_state=42)
        
        r_squared= []
        rmse= []
        mae= []
        
        #Hata ve doğruluk oranlarını bir tablo haline getirmek için bir dataframe oluşturuyorum
        result=pd.DataFrame(columns=['R_Squared','RMSE','MAE'],index=algo_names)
        
        
        for algo in algos:
            p=algo.fit(x_train,y_train).predict(x_test)
            r_squared.append(r2_score(y_test,p))
            rmse.append(mean_squared_error(y_test,p)**.5)
            mae.append(mean_absolute_error(y_test,p))
        
            

        #result adlı tabloya doğruluk ve hata oranlarımı yerleştiriyorum
        result.R_Squared=r_squared
        result.RMSE=rmse
        result.MAE=mae
        
       #oluşturduğum result tablosunu doğruluk oranına (r2_score) göre sıralayıp dönüyor
        rtable=result.sort_values('R_Squared',ascending=False)
        return rtable
```
